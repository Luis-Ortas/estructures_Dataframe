### Añadidmos columnas
**Retraso:**
* 0 no hay retraso
* 1 Si hay retraso.



```python
## definimos el retraso como mayor a 30 minutos. También los Desviados y cancelados aparecen como retraso.
def retraso_bin(x):
    
    '''Se define como retraso encima de 30 min más 
    los vuelos desviados y candelados'''
    
    if x>=30.0:
        retraso=1
    
    elif x<30:
        retraso=0
    else:
        retraso=1
        
  
    return retraso

df_vuelo["Retraso"]=df['ArrDelay'].apply(retraso_bin)

df_vuelo.head()
```

```python
num_retraso=df_vuelo['Retraso'].value_counts(normalize=True)
num_retraso.head()
```

```python
 ## cambiamos el formato de lo
print("En el 2018 un {:2.2%}  de los vuelos han tenido retraso".format(num_retraso[0]))
```

```python
df_vuelo['Velocidad']=df_vuelo['Distance']/(df_vuelo['AirTime']/60)
df_vuelo['Velocidad'].round(2)
```

```python
## Eliminamo los valores infinitos. Lo sustitumos por nan

df_vuelo.replace([np.inf, -np.inf], np.nan, inplace=True)
```

### Analisis de los retrasos
#### Retrasos por:
* Dia Total

```python
df_vuelo.groupby('DayOfWeek').agg({'Retraso':sum,'FlightNum':len}).sort_values(by='Retraso',ascending=False)
```

* Dia %

```python
dia=df_vuelo.groupby(['DayOfWeek','Retraso']).agg(len)

dia['Porcentaje']=df_vuelo.groupby('DayOfWeek')["Retraso"].value_counts(normalize=True)

dia.reset_index(level=1, inplace=True)
#prueba['Cuenta_Retraso']=prueba.index
dia_sim=dia[dia['Retraso']==1].sort_values(by='Porcentaje',ascending=False)
pd.set_option('display.float_format', '{:.2%}'.format) ## cambiamos el formato de los
dia_sim[['Retraso','Porcentaje']]
```

```python
pd.reset_option('display.float_format')
```

* Mes Total:

```python
df_vuelo.groupby('Month').agg({'Retraso':sum,'FlightNum':len}).sort_values(by='Retraso',ascending=False)
```

* Mes %

```python
month=df_vuelo.groupby(['Month','Retraso']).agg(len)

month['Porcentaje']=df_vuelo.groupby('Month')["Retraso"].value_counts(normalize=True)

month.reset_index(level=1, inplace=True)
#prueba['Cuenta_Retraso']=prueba.index
month_sim=month[month['Retraso']==1].sort_values(by='Porcentaje',ascending=False)
pd.set_option('display.float_format', '{:.2%}'.format) ## cambiamos el formato de los
month_sim[['Porcentaje']]
```

* Compañía Total

```python
df_vuelo.groupby('UniqueCarrier').agg({'Retraso':sum,'FlightNum':len}).sort_values(by='Retraso',ascending=False)
```

* Compañia %

```python
cia=df_vuelo.groupby(['UniqueCarrier','Retraso']).agg(len)

cia['Porcentaje']=df_vuelo.groupby('UniqueCarrier')["Retraso"].value_counts(normalize=True)

cia.reset_index(level=1, inplace=True)
#prueba['Cuenta_Retraso']=prueba.index
cia_sim=cia[cia['Retraso']==1].sort_values(by='Porcentaje',ascending=False)
pd.set_option('display.float_format', '{:.2%}'.format) ## cambiamos el formato de los
cia_sim[['Porcentaje']]
```

```python
cia_total=df_vuelo.groupby('UniqueCarrier').agg({'Retraso':sum,'FlightNum':len})
cia=df_vuelo.groupby(['UniqueCarrier','Retraso']).agg(len)

cia['Porcentaje']=df_vuelo.groupby('UniqueCarrier')["Retraso"].value_counts(normalize=True)

cia.reset_index(level=1, inplace=True)
#prueba['Cuenta_Retraso']=prueba.index
cia_sim=cia[cia['Retraso']==1].sort_values(by='Porcentaje',ascending=False)
pd.set_option('display.float_format', '{:.2%}'.format) ## cambiamos el formato de los
cia_total['Porcentaje']=cia_sim['Porcentaje']
cia_total.rename(columns={'Retraso':'N Vuelos Retrasados','FlightNum':'Total Vuelos','Porcentaje':'Porcentaje_Retrasos'}, inplace=True)
cia_total.sort_values(by='Porcentaje_Retrasos',ascending=False)
```

```python
vuel_num=df_vuelo.groupby(['FlightNum','Retraso']).agg(len)

vuel_num['Porcentaje']=df_vuelo.groupby('FlightNum')["Retraso"].value_counts(normalize=True)

vuel_num.reset_index(level=1, inplace=True)
#prueba['Cuenta_Retraso']=prueba.index
vuel_num_sim=vuel_num[vuel_num['Retraso']==1].sort_values(by='Porcentaje',ascending=False)
pd.set_option('display.float_format', '{:.2%}'.format) ## cambiamos el formato de los
vuel_num_sim[['Porcentaje']]
```
