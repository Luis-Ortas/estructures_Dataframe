### Creamos Columnas

```python
def funcion_horasolo(x):
    
    if pd.notnull(x):
        
        x=int(x)
        if x==2400:
            x=0
       
    
   
        hora= int(x/100)
        hora=int(hora)
    
        return hora
    else:
        return 25 # si ponemos None queda como float toda la columna, los no nulos.(estudiar)
    
    
    
def funcion_minutosolo(x):
    if pd.notnull(x):
        
        x=int(x)
        if x==2400:
            x=0
    
   
        minuto= int(x%100)
        minuto=int(minuto)
    
        return int(minuto)
    
    else:
        return 61 # si ponemos None queda como float toda la columna, los no nulos.
```

```python
# Creamos columna 'DepTime'fecha hora
df["Horas"]=df['DepTime'].apply(funcion_horasolo)
df["Minutos"]=df['DepTime'].apply(funcion_minutosolo)

#Construimos la columna
df['DepTime_FH']=df.apply(
    lambda row: datetime(row['Year'],row['Month'],row['DayofMonth'],row['Horas'],row['Minutos']),axis=1
)
df.drop('Horas', axis=1,inplace=True)
df.drop('Minutos', axis=1,inplace=True)
```

```python
# Creamos columna 'CRSDepTime'fecha hora
df["Horas"]=df['CRSDepTime'].apply(funcion_horasolo)
df["Minutos"]=df['CRSDepTime'].apply(funcion_minutosolo)

df['CRSDepTime_FH']=df.apply(
    lambda row: datetime(row['Year'],row['Month'],row['DayofMonth'],row['Horas'],row['Minutos']),axis=1
)
df.drop('Horas', axis=1,inplace=True)
df.drop('Minutos', axis=1,inplace=True)
```

```python
# Creamos columna 'ArrTime'fecha hora
df["Horas"]=df['ArrTime'].apply(funcion_horasolo)
df["Minutos"]=df['ArrTime'].apply(funcion_minutosolo)

df_nonulos=df[df['ArrTime'].notnull()] # lo aplicaremos unicamente a los registros que DepTime no sea NaN.

df['ArrTime_FH']=df_nonulos.apply(
   lambda row: datetime(row['Year'],row['Month'],row['DayofMonth'],row['Horas'],row['Minutos']),axis=1
)
df.drop('Horas', axis=1,inplace=True)
df.drop('Minutos', axis=1,inplace=True)
```

```python
# Creamos columna 'CRSArrTime'fecha hora
df["Horas"]=df['CRSArrTime'].apply(funcion_horasolo)
df["Minutos"]=df['CRSArrTime'].apply(funcion_minutosolo)
df['CRSArrTime_FH']=df.apply(
    lambda row: datetime(row['Year'],row['Month'],row['DayofMonth'],row['Horas'],row['Minutos']),axis=1
)
df.drop('Horas', axis=1,inplace=True)
df.drop('Minutos', axis=1,inplace=True)
```

```python
# calculamos retraso
df['Retraso_Llegada']=df['ArrTime_FH']-df['CRSArrTime_FH']
df['Retraso_Salida']=df['DepTime_FH']-df['CRSDepTime_FH']
```

```python
def total_seconds(x):
    if pd.notnull(x):
        secundos=x.total_seconds()
        minutos=secundos/60
        return minutos
    else:
        return None
```

```python
#Creamos columnas retrasos en minutos.(float)
df['Retraso_llegas_min']=df['Retraso_Llegada'].apply(total_seconds)
df['Retraso_salida_min']=df['Retraso_Salida'].apply(total_seconds)
```
